## Exercise 1: Simple Area Calculator

Write a function that calculates the area of a rectangle.
- Function name: `rectangleArea`
- Parameters: `length`, `width`
- Return: The area (length * width)
- Include a docstring

```python
def rectangleArea(length, width):
    """Calculate the area of a rectangle given length and width."""
    return length * width

# Test your function
print(rectangleArea(5, 10))  # Should print 50
print(rectangleArea(3, 4))   # Should print 12
```

**Instructor Note**: Simple function with two parameters and a return statement. Good baseline for understanding function structure.

## Exercise 2: Circle Area Function

Write a function that calculates the area of a circle.
- Function name: `circleArea`
- Parameter: `radius`
- Return: The area (pi * radius^2)
- Use the math module for pi
- Include a docstring

```python
import math

def circleArea(radius):
    """Calculate the area of a circle given the radius."""
    return math.pi * radius ** 2

# Test your function
print(f"Area of circle with radius 5: {circleArea(5):.2f}")  # Should be ~78.54
print(f"Area of circle with radius 10: {circleArea(10):.2f}") # Should be ~314.16
```

**Instructor Note**: Introduces using the math module inside functions. Note that we import math before defining the function.

## Exercise 3: BMI Calculator

Write a function that calculates Body Mass Index (BMI).
- Function name: `calculateBMI`
- Parameters: `weight` (pounds), `height` (inches)
- Formula: BMI = (weight / height^2) * 703
- Return: The BMI value
- Include a detailed docstring (parameters, return value, notes)

```python
def calculateBMI(weight, height):
    """
    Calculate Body Mass Index (BMI).
    
    Parameters:
    - weight: Weight in pounds
    - height: Height in inches
    
    Returns:
    - BMI value (float)
    
    Formula: BMI = (weight / height^2) * 703
    """
    bmi = (weight / (height ** 2)) * 703
    return bmi

# Test your function
myBMI = calculateBMI(180, 70)
print(f"BMI for 180 lbs, 70 inches: {myBMI:.1f}")  # Should be ~25.8
```

**Instructor Note**: Demonstrates detailed docstrings with parameter descriptions. Formula is important — 703 is the conversion factor for pounds and inches.

## Exercise 4: Letter Grade Function

Write a function that converts a numeric score to a letter grade.
- Function name: `getLetterGrade`
- Parameter: `score` (0-100)
- Grading scale:
  - A: 90-100
  - B: 80-89
  - C: 70-79
  - D: 60-69
  - F: Below 60
- Return: The letter grade as a string
- Include a docstring

```python
def getLetterGrade(score):
    """
    Convert a numeric score to a letter grade.
    
    Parameters:
    - score: Numeric score (0-100)
    
    Returns:
    - Letter grade (A, B, C, D, or F)
    """
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Test your function
print(getLetterGrade(95))  # Should print A
print(getLetterGrade(85))  # Should print B
print(getLetterGrade(75))  # Should print C
print(getLetterGrade(65))  # Should print D
print(getLetterGrade(55))  # Should print F
```

**Instructor Note**: Classic if-elif-else chain. Returns a string. Note that `>=` is important — 90 is an A, not a B.

## Exercise 5: Function with Default Parameter

Write a function that calculates the price after tax.
- Function name: `priceWithTax`
- Parameters: `price`, `taxRate=0.06` (6% default tax)
- Return: The total price (price + price * taxRate)
- Include a docstring

```python
def priceWithTax(price, taxRate=0.06):
    """
    Calculate the total price after tax.
    
    Parameters:
    - price: Original price
    - taxRate: Tax rate (default 0.06 or 6%)
    
    Returns:
    - Total price after tax
    """
    return price + (price * taxRate)

# Test your function
print(f"Price with default tax: ${priceWithTax(100):.2f}")      # Should be ~106.00
print(f"Price with 10% tax: ${priceWithTax(100, 0.10):.2f}")    # Should be 110.00
```

**Instructor Note**: Demonstrates default parameters. When called with one argument, it uses the default tax rate. When called with two, it uses the provided rate.

## Exercise 6: Temperature Converter

Write a function that converts Fahrenheit to Celsius.
- Function name: `fahrenheitToCelsius`
- Parameter: `fahrenheit`
- Formula: celsius = (fahrenheit - 32) * 5 / 9
- Return: The temperature in Celsius
- Include a docstring

```python
def fahrenheitToCelsius(fahrenheit):
    """
    Convert temperature from Fahrenheit to Celsius.
    
    Parameters:
    - fahrenheit: Temperature in Fahrenheit
    
    Returns:
    - Temperature in Celsius
    """
    celsius = (fahrenheit - 32) * 5 / 9
    return celsius

# Test your function
print(f"32°F = {fahrenheitToCelsius(32):.1f}°C")  # Should be 0.0
print(f"68°F = {fahrenheitToCelsius(68):.1f}°C")  # Should be ~20.0
print(f"212°F = {fahrenheitToCelsius(212):.1f}°C") # Should be 100.0
```

**Instructor Note**: Standard temperature conversion formula. Good for practicing arithmetic in functions.

## Exercise 7: String Greeting Function

Write a function that creates a greeting message.
- Function name: `buildGreeting`
- Parameters: `firstName`, `lastName`, `title="Mr."`
- Return: A formatted string like "Hello, Mr. John Smith!"
- Include a docstring

```python
def buildGreeting(firstName, lastName, title="Mr."):
    """
    Build a greeting message with a person's name and title.
    
    Parameters:
    - firstName: Person's first name
    - lastName: Person's last name
    - title: Title (default "Mr.")
    
    Returns:
    - Formatted greeting string
    """
    return f"Hello, {title} {firstName} {lastName}!"

# Test your function
print(buildGreeting("John", "Smith"))                 # Should be "Hello, Mr. John Smith!"
print(buildGreeting("Jane", "Doe", "Ms."))             # Should be "Hello, Ms. Jane Doe!"
```

**Instructor Note**: String formatting with f-strings inside a function. Default parameter with a string value.

## Exercise 8: Challenge - Absolute Value Function

Write a function that returns the absolute value of a number (without using Python's built-in `abs()`).
- Function name: `myAbsoluteValue`
- Parameter: `number`
- Return: The absolute value
- Use an if statement
- Include a docstring

```python
def myAbsoluteValue(number):
    """
    Return the absolute value of a number.
    
    Parameters:
    - number: Any numeric value
    
    Returns:
    - Absolute value (always positive)
    """
    if number < 0:
        return -number
    else:
        return number

# Test your function
print(myAbsoluteValue(5))    # Should be 5
print(myAbsoluteValue(-5))   # Should be 5
print(myAbsoluteValue(0))    # Should be 0
print(myAbsoluteValue(-3.7)) # Should be 3.7
```

**Instructor Note**: Challenge exercise. Shows how to use logic to implement mathematical concepts. Alternative approach: `return number if number >= 0 else -number`
