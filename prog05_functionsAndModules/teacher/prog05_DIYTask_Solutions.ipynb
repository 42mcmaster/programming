## Part 1: Roll Dice Function

```python
from random import randint

def rollDice():
    """Roll two dice and return the sum."""
    die1 = randint(1, 6)
    die2 = randint(1, 6)
    return die1 + die2

# Test
for i in range(5):
    print(f"Roll {i+1}: {rollDice()}")
```

**Instructor Note**: Simple function using random.randint. Returns sum of two dice.

## Part 2: Check First Roll Function

```python
def checkFirstRoll(roll):
    """
    Check the result of the first roll.
    Returns: 'win', 'lose', or the point number.
    """
    if roll == 7 or roll == 11:
        return "win"
    elif roll == 2 or roll == 3 or roll == 12:
        return "lose"
    else:
        return roll  # Return the point

# Test
print(checkFirstRoll(7))   # Should be "win"
print(checkFirstRoll(2))   # Should be "lose"
print(checkFirstRoll(6))   # Should be 6
```

**Instructor Note**: Returns different types (string or int). In Python, this is fine.

## Part 3: Check Point Roll Function

```python
def checkPointRoll(roll, point):
    """
    Check if point roll wins, loses, or continues.
    Returns: 'win', 'lose', or 'continue'.
    """
    if roll == point:
        return "win"
    elif roll == 7:
        return "lose"
    else:
        return "continue"

# Test
print(checkPointRoll(6, 6))   # Point was 6, rolled 6: "win"
print(checkPointRoll(7, 6))   # Rolled 7: "lose"
print(checkPointRoll(4, 6))   # Rolled 4: "continue"
```

**Instructor Note**: Checks game state during point phase. Clean logic.

## Part 4: Play One Game Function

```python
def playGame():
    """
    Play one complete game.
    """
    print("\n--- Starting new game ---")
    
    # First roll
    roll = rollDice()
    print(f"First roll: {roll}")
    
    result = checkFirstRoll(roll)
    
    if result == "win":
        print("Winner! You rolled 7 or 11!")
        return
    elif result == "lose":
        print("Craps! You rolled 2, 3, or 12!")
        return
    
    # Point phase
    point = result
    print(f"Your point is {point}. Keep rolling...")
    
    while True:
        roll = rollDice()
        print(f"Rolled: {roll}")
        
        result = checkPointRoll(roll, point)
        
        if result == "win":
            print(f"Winner! You rolled your point ({point})!")
            return
        elif result == "lose":
            print(f"Lost! You rolled a 7!")
            return
        # else: continue

# Test
playGame()
```

**Instructor Note**: Orchestrates the full game. Calls helper functions. Uses loops and conditionals well.

## Part 5: Main Program

```python
# Main program
print("Welcome to the Dice Game!")
playAgain = "yes"

while playAgain.lower() == "yes":
    playGame()
    playAgain = input("\nPlay again? (yes/no): ")

print("\nThanks for playing!")
```

**Instructor Note**: Main loop allows multiple games. Clean program structure.
