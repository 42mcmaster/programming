# prog05c_Task: Multi-Function Program — Simple Calculator

## Part 1: Create Individual Operation Functions

```python
def add(num1, num2):
    """Add two numbers and return the result."""
    return num1 + num2

def subtract(num1, num2):
    """Subtract num2 from num1 and return the result."""
    return num1 - num2

def multiply(num1, num2):
    """Multiply two numbers and return the result."""
    return num1 * num2

def divide(num1, num2):
    """Divide num1 by num2. Returns error message if dividing by zero."""
    if num2 == 0:
        return "Error: Cannot divide by zero!"
    return num1 / num2

# Test the functions
print(f"add(10, 5) = {add(10, 5)}")
print(f"subtract(10, 5) = {subtract(10, 5)}")
print(f"multiply(10, 5) = {multiply(10, 5)}")
print(f"divide(10, 5) = {divide(10, 5)}")
```

**Instructor Note**: Each function is simple and focused. The divide function checks for zero division.

## Part 2: Create a Menu Display Function

```python
def displayMenu():
    """Display the calculator menu options."""
    print("===== Simple Calculator =====")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")
    print("=============================")

displayMenu()
```

**Instructor Note**: Simple display function. It does one thing well.

## Part 3: Create a Function That Performs Calculation

```python
def performCalculation(operation, num1, num2):
    """
    Perform the specified calculation.
    
    Parameters:
    - operation: 1 (add), 2 (subtract), 3 (multiply), 4 (divide)
    - num1: First operand
    - num2: Second operand
    
    Returns:
    - Result of the operation
    """
    if operation == 1:
        return add(num1, num2)
    elif operation == 2:
        return subtract(num1, num2)
    elif operation == 3:
        return multiply(num1, num2)
    elif operation == 4:
        return divide(num1, num2)
    else:
        return "Invalid operation."

# Test it
result = performCalculation(1, 10, 5)
print(f"Operation 1 with 10, 5: {result}")

result = performCalculation(4, 20, 4)
print(f"Operation 4 with 20, 4: {result}")
```

**Instructor Note**: This function calls other functions. Key pattern in modular programming.

## Part 4: Create the Main Calculator Program

```python
def runCalculator():
    """
    Main calculator program. Loop until user exits.
    """
    while True:
        displayMenu()
        operation = int(input("Choose operation (1-5): "))
        if operation == 5:
            print("Thank you for using the calculator! Goodbye.")
            break
        if operation < 1 or operation > 4:
            print("Invalid operation. Please try again.")
            print()
            continue
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        result = performCalculation(operation, num1, num2)
        print(f"Result: {result}")
        print()

# Uncomment to test
# runCalculator()
```

**Instructor Note**: Complete program using while loop. Handles user input and error cases.
