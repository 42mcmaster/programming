## Project: Number Guessing Game — COMPLETE SOLUTION

Create a complete number guessing game where:

1. The computer picks a random number between 1 and 100
2. The player guesses the number
3. The program gives hints: "Too high" or "Too low"
4. Count the number of attempts
5. When the player guesses correctly, show the number of attempts and ask if they want to play again
6. Keep playing until the user says no

### Step 1: Import random and understand random.randint()

```python
# This generates a random number between 1 and 100 (inclusive)
import random

secretNumber = random.randint(1, 100)
print(f"Secret number for testing: {secretNumber}")
```

---
### Step 2: Build the basic guessing logic (single game)

```python
# SOLUTION Step 2 — Single Game
import random

# Generate a random number between 1 and 100
secretNumber = random.randint(1, 100)

# Counter for attempts
attempts = 0

# Game loop
while True:
    # Get and validate user input
    while True:
        guessInput = input("Guess the number (1-100): ")
        # Check if input is a valid number
        if guessInput.isdigit():
            guess = int(guessInput)
            if 1 <= guess <= 100:
                break  # Valid input
        print("Invalid input. Please enter a number between 1 and 100.")

    # Increment attempt counter
    attempts = attempts + 1

    # Check guess and give feedback
    if guess == secretNumber:
        print(f"Correct! You got it in {attempts} attempt(s)!")
        break  # Exit game loop
    elif guess < secretNumber:
        print("Too low! Try again.")
    else:  # guess > secretNumber
        print("Too high! Try again.")
```

---
### Step 3: Add replay functionality (COMPLETE SOLUTION)

```python
# SOLUTION Step 3 — Complete Game with Replay
import random

# Welcome message
print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 100.\n")

# Outer loop — play multiple games (sentinel loop)
while True:
    # Generate a random number between 1 and 100
    secretNumber = random.randint(1, 100)

    # Counter for attempts
    attempts = 0

    # Inner loop — play one game
    while True:
        # Get and validate user input
        while True:
            guessInput = input("Guess the number: ")
            # Check if input is a valid number
            if guessInput.isdigit():
                guess = int(guessInput)
                if 1 <= guess <= 100:
                    break  # Valid input
            print("Invalid input. Please enter a number between 1 and 100.")

        # Increment attempt counter
        attempts = attempts + 1

        # Check guess and give feedback
        if guess == secretNumber:
            print(f"Correct! You got it in {attempts} attempt(s)!")
            break  # Exit game loop (one game finished)
        elif guess < secretNumber:
            print("Too low! Try again.")
        else:  # guess > secretNumber
            print("Too high! Try again.")

    # Ask if player wants to play again
    print()  # Blank line for readability
    while True:
        playAgain = input("Do you want to play again? (yes/no): ").lower()
        if playAgain in ["yes", "no", "y", "n"]:
            break
        print("Please enter 'yes' or 'no'.")

    # Check if user wants to continue
    if playAgain in ["no", "n"]:
        print("\nThanks for playing!")
        break  # Exit outer loop (stop playing)
```
