## Exercise 1: Grade Checker

Write a program that takes a test score (0-100) and prints the letter grade:
- 90-100: A
- 80-89: B
- 70-79: C
- 60-69: D
- Below 60: F

```python
# SOLUTION 1
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}")
print(f"Grade: {grade}")
```

## Exercise 2: Number Classifier

Write a program that classifies a number:
- If positive: print "Positive"
- If negative: print "Negative"
- If zero: print "Zero"

Also check if it's even or odd (positive or negative).

```python
# SOLUTION 2
number = -7

# Check if positive, negative, or zero
if number > 0:
    print(f"{number} is Positive")
elif number < 0:
    print(f"{number} is Negative")
else:
    print(f"{number} is Zero")

# Check if even or odd (works for positive and negative)
# % is modulo operator — num % 2 == 0 means even
if number != 0:  # Zero is even but let's skip the parity check for it
    if number % 2 == 0:
        print(f"{number} is Even")
    else:
        print(f"{number} is Odd")
```

Create a simple menu system:

## Exercise 4: Eligibility Checker

Check if someone is eligible for a scholarship:
- Must have GPA >= 3.5 AND SAT score >= 1200
- OR be a first-generation college student with GPA >= 3.0

Use logical operators (and, or).

```python
# SOLUTION 4
gpa = 3.6
satScore = 1150
isFirstGeneration = True

# Eligible if:
# (GPA >= 3.5 AND SAT >= 1200) OR (first-gen AND GPA >= 3.0)
if (gpa >= 3.5 and satScore >= 1200) or (isFirstGeneration and gpa >= 3.0):
    print("Eligible for scholarship!")
else:
    print("Not eligible for scholarship.")

# Show details
print(f"GPA: {gpa}")
print(f"SAT: {satScore}")
print(f"First-generation: {isFirstGeneration}")
```

## Exercise 5: Age-Based Access

Check access based on age:
- If age < 13: "Kids section"
- If 13 <= age < 18: "Teens section"
- If 18 <= age < 65: "Adult section"
- If age >= 65: "Senior section"

Also check if they have a valid ID (boolean).

```python
# SOLUTION 5
age = 25
hasValidID = True

# First check ID (for adults)
if age >= 18 and not hasValidID:
    print("You need a valid ID to enter.")
else:
    # Then assign section based on age
    if age < 13:
        section = "Kids section"
    elif age < 18:
        section = "Teens section"
    elif age < 65:
        section = "Adult section"
    else:
        section = "Senior section"

    print(f"Age: {age}")
    print(f"Valid ID: {hasValidID}")
    print(f"Access: {section}")
```

## Exercise 6: Temperature Recommendation

Give a clothing recommendation based on temperature:
- Below 32°F: "Wear a heavy coat"
- 32-50°F: "Wear a jacket"
- 51-70°F: "Wear a sweater"
- 71-85°F: "Wear light clothing"
- Above 85°F: "Wear very light clothing"

If it's raining, add "and bring an umbrella!"

```python
# SOLUTION 6
temperature = 65
isRaining = False

# Determine clothing recommendation
if temperature < 32:
    clothing = "Wear a heavy coat"
elif temperature < 51:
    clothing = "Wear a jacket"
elif temperature < 71:
    clothing = "Wear a sweater"
elif temperature <= 85:
    clothing = "Wear light clothing"
else:
    clothing = "Wear very light clothing"

# Add rain recommendation
if isRaining:
    clothing = clothing + " and bring an umbrella!"

print(f"Temperature: {temperature}°F")
print(f"Raining: {isRaining}")
print(clothing)
```

## Exercise 7: Movie Recommendation

Recommend a movie genre based on mood and preferences:
- If mood is "happy" and likes action: "Action Comedy"
- If mood is "sad": "Drama"
- If mood is "scared": "Horror"
- If likes action: "Action"

## Challenge: Discount Calculator

Calculate a discount based on purchase amount and customer type:
- If purchase >= $100 and is member: 20% discount
- If purchase >= $100 and NOT member: 10% discount
- If purchase >= $50 and is member: 10% discount
- If purchase >= $50 and NOT member: 5% discount
- Otherwise: no discount

Print the original price, discount percentage, and final price.

```python
# SOLUTION Challenge
purchaseAmount = 125.00
isMember = True

# Determine discount percentage
if purchaseAmount >= 100 and isMember:
    discountPercent = 20
elif purchaseAmount >= 100 and not isMember:
    discountPercent = 10
elif purchaseAmount >= 50 and isMember:
    discountPercent = 10
elif purchaseAmount >= 50 and not isMember:
    discountPercent = 5
else:
    discountPercent = 0

# Calculate discount amount and final price
discountAmount = purchaseAmount * (discountPercent / 100)
finalPrice = purchaseAmount - discountAmount

# Print results
print(f"Original price: ${purchaseAmount:.2f}")
print(f"Member: {isMember}")
print(f"Discount: {discountPercent}%")
print(f"Discount amount: ${discountAmount:.2f}")
print(f"Final price: ${finalPrice:.2f}")
```
