## Part 1: Error Handling Fundamentals

### Try/Except Block Structure

In a BPA contest, your code cannot crash. Bad input will happen. Use try/except to handle errors gracefully.

**Pattern:**
```python
try:
    # code that might cause an error
except SpecificException:
    # handle that specific error
```

### Example 1: Handling ValueError

When user input is not the expected type, int() raises ValueError.

````python
def getPositiveInteger():
    while True:
        try:
            user_input = input('Enter a positive integer: ')
            num = int(user_input)
            if num > 0:
                return num
            else:
                print('Please enter a positive number.')
        except ValueError:
            print('Invalid input. Please enter a valid integer.')

# result = getPositiveInteger()
# print(f'You entered: {result}')
````

### Example 2: Handling Multiple Exceptions

````python
def safeDivide(a, b):
    try:
        numerator = int(a)
        denominator = int(b)
        result = numerator / denominator
        return result
    except ValueError:
        print('Error: Both inputs must be valid integers.')
        return None
    except ZeroDivisionError:
        print('Error: Cannot divide by zero.')
        return None

print(safeDivide(10, 2))
print(safeDivide(10, 0))
print(safeDivide('abc', 2))
````

## Summary

**Key Takeaways:**
