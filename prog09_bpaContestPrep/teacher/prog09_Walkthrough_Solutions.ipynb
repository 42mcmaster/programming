## Part 1: Error Handling Fundamentals

### Example 1: Handling ValueError

```python
def getPositiveInteger():
    while True:
        try:
            user_input = input('Enter a positive integer: ')
            num = int(user_input)
            if num > 0:
                return num
            else:
                print('Please enter a positive number.')
        except ValueError:
            print('Invalid input. Please enter a valid integer.')

# Instructor Note: Input validation with try/except.
```
