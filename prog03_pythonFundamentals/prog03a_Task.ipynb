## Key Takeaway

Notice how much simpler Python is:
- No `#include`, no `main()`, no `return 0`
- No type declarations (Python figures it out)
- No `scanf()` complications (just `input()`)
- F-strings replace `printf()` format strings
- No semicolons, no braces — indentation does the job

You can now focus on the **logic** of your programs, not the **syntax** of the language.
