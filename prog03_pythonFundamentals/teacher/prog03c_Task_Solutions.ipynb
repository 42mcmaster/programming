## Program 1: Temperature Converter

```python
# Program: Convert Celsius to Fahrenheit
# Author: Ryan McMaster

# Input
celsius = float(input("Enter temperature in Celsius: "))

# Processing
fahrenheit = (celsius * 9 / 5) + 32

# Output
print(f"{celsius}°C = {fahrenheit:.0f}°F")
```

## Program 2: Tip Calculator

```python
# Program: Calculate tip and total bill
# Author: Ryan McMaster

DEFAULT_TIP_PERCENT = 0.18

# Input
billAmount = float(input("Enter bill amount: $"))
tipPercent = float(input("Enter tip percentage (0-100): ")) / 100

# Processing
tipAmount = billAmount * tipPercent
totalAmount = billAmount + tipAmount

# Output
print(f"Bill: ${billAmount:.2f}")
print(f"Tip: ${tipAmount:.2f}")
print(f"Total: ${totalAmount:.2f}")
```

## Program 3: Simple Payroll Calculator

```python
# Program: Calculate gross pay, tax, and net pay
# Author: Ryan McMaster

TAX_RATE = 0.15

# Input
hourlyWage = float(input("Enter hourly wage: $"))
hoursWorked = float(input("Enter hours worked: "))

# Processing
grossPay = hourlyWage * hoursWorked
taxAmount = grossPay * TAX_RATE
netPay = grossPay - taxAmount

# Output
print("---")
print(f"Gross Pay: ${grossPay:.2f}")
print(f"Tax (15%): ${taxAmount:.2f}")
print(f"Net Pay: ${netPay:.2f}")
```

## Program 4: Area of a Rectangle

```python
# Program: Calculate area and perimeter of a rectangle
# Author: Ryan McMaster

# Input
length = float(input("Enter length: "))
width = float(input("Enter width: "))

# Processing
area = length * width
perimeter = 2 * (length + width)

# Output
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
```

## Program 5: GPA Calculator

```python
# Program: Calculate average test score
# Author: Ryan McMaster

# Input
studentName = input("Enter student name: ")
test1 = float(input("Enter test 1 score: "))
test2 = float(input("Enter test 2 score: "))
test3 = float(input("Enter test 3 score: "))

# Processing
average = (test1 + test2 + test3) / 3

# Output
print("---")
print(f"Student: {studentName}")
print(f"Average: {average:.2f}")
```
