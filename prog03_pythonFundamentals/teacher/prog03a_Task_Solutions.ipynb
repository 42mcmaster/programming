## Exercise 3: Declare and Print Variables

```python
age = 17
height = 5.8
initial = 'R'

print(f"Age: {age}, Height: {height}, Initial: {initial}")
```

**Instructor Note:** In Python, single characters are just strings. No `char` type like C. The f-string automatically formats everything nicely without needing format specifiers like `%.1f`.

## Exercise 4: Arithmetic and Print

```python
a = 10
b = 3

print(f"Sum: {a + b}")
print(f"Product: {a * b}")
print(f"Division: {a / b}")
```

**Instructor Note:** Key point: in C, you'd need `(float)a / b` to get a float division. In Python, `/` always returns a float automatically. No casting needed.
