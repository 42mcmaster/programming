## Complete Program: Personal Info Card

```python
# Program: Personal Info Card
# Author: Student Name
# Description: Collects user info and displays a formatted card

CURRENT_YEAR = 2024
YEARS_TO_GRADUATE = 4

# Input
fullName = input("Enter your full name: ")
age = int(input("Enter your age: "))
favoriteHobby = input("Enter your favorite hobby: ")
numberOfPets = int(input("How many pets do you have? "))
currentAverage = float(input("Enter your current class average: "))

# Processing
graduationYear = CURRENT_YEAR + YEARS_TO_GRADUATE - age
gradePercentage = currentAverage * 25
petsPerYearOfAge = numberOfPets / age

# Output
print("========================================")
print("PERSONAL INFO CARD")
print("========================================")
print(f"Name: {fullName}")
print(f"Age: {age}")
print(f"Favorite Hobby: {favoriteHobby}")
print(f"Number of Pets: {numberOfPets}")
print()
print(f"Current Average: {currentAverage}")
print(f"Grade Percentage: {gradePercentage:.2f}%")
print()
print(f"Expected Graduation Year: {int(graduationYear)}")
print(f"Pets per Year of Age: {petsPerYearOfAge:.2f}")
print("========================================")
```
