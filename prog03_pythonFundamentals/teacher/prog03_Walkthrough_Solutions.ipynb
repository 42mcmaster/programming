### Solution: Calculate and Compare Division

```python
# Compare / and //
resultFloat = 15 / 2
resultInt = 15 // 2

print(f"15 / 2 = {resultFloat}")
print(f"Type: {type(resultFloat)}")
print()
print(f"15 // 2 = {resultInt}")
print(f"Type: {type(resultInt)}")
```

**Instructor Note:** `/` returns float, `//` returns int. Different from C.

## Complete Reference Examples

```python
# EXAMPLE: Temperature Converter
celsiusInput = input("Enter temperature in Celsius: ")
celsius = float(celsiusInput)
fahrenheit = (celsius * 9 / 5) + 32
print(f"{celsius}°C = {fahrenheit:.1f}°F")
```

```python
# EXAMPLE: Tip Calculator
TIP_PERCENT = 0.18

billAmount = float(input("Enter your bill amount: $"))
tipAmount = billAmount * TIP_PERCENT
totalAmount = billAmount + tipAmount

print(f"Bill: ${billAmount:.2f}")
print(f"Tip (18%): ${tipAmount:.2f}")
print(f"Total: ${totalAmount:.2f}")
```
