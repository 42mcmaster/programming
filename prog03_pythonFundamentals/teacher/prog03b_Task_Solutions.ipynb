## Exercise 4: Division Differences

```python
a = 10
b = 3

regularDivision = a / b
integerDivision = a // b
remainder = a % b

print(f"a / b = {regularDivision} (type: {type(regularDivision)})")
print(f"a // b = {integerDivision} (type: {type(integerDivision)})")
print(f"a % b = {remainder}")
```
