## Project Overview

Create a class `TextAnalyzer` that takes a paragraph of text and computes:
1. **Word Count** — total number of words
2. **Sentence Count** — total number of sentences
3. **Average Word Length** — mean characters per word
4. **Most Common Word** — the word that appears most frequently
5. **Longest Word** — the word with the most characters

---

## Requirements

- Create a class called `TextAnalyzer`
- Constructor takes text as parameter
- Implement methods:
  - `wordCount()` → returns int
  - `sentenceCount()` → returns int
  - `averageWordLength()` → returns float (rounded to 2 decimal places)
  - `mostCommonWord()` → returns str (the word, all lowercase)
  - `longestWord()` → returns str (the word, preserving original case)
  - `analyze()` → prints a formatted report

- **Rules**:
  - Remove punctuation before analyzing
  - Treat uppercase/lowercase as the same word (lowercase for comparison)
  - Sentences end with `.`, `!`, or `?`
  - Words are separated by spaces
