## Exercise 1: Word Frequency Counter

```python
def countWordFrequency(text):
    # Instructor Note: Clean, split, count
    import string
    
    # Remove punctuation and lowercase
    cleaned = text.lower()
    for char in string.punctuation:
        cleaned = cleaned.replace(char, "")
    
    # Split into words
    words = cleaned.split()
    
    # Count frequencies
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    
    return frequency

# Test
sample = "Hello world! Hello Python. Python is fun."
result = countWordFrequency(sample)
print(result)

sample2 = "The quick brown fox jumps over the lazy dog."
print(countWordFrequency(sample2))
```

## Exercise 2: Parse CSV Data

```python
def parseCSV(csvString):
    # Instructor Note: Split by newline, then split each row by comma
    lines = csvString.strip().split('\n')
    result = []
    
    for line in lines:
        row = line.split(',')
        result.append(row)
    
    return result

# Test
csvData = """name,age,city
Alice,30,Boston
Bob,25,NYC
Carol,28,LA"""

result = parseCSV(csvData)
print("Parsed CSV:")
for row in result:
    print(row)

# Can access like 2D array
print(f"\nFirst name from row 1: {result[1][0]}")
print(f"Age from row 2: {result[2][1]}")
```

## Exercise 3: Grade Book

```python
class GradeBook:
    # Instructor Note: Store students as list of dicts
    def __init__(self):
        self.students = []  # List of {name, grades}
    
    def addStudent(self, name, grades):
        self.students.append({'name': name, 'grades': grades})
    
    def getAverage(self, name):
        for student in self.students:
            if student['name'] == name:
                if len(student['grades']) == 0:
                    return 0
                total = 0
                for grade in student['grades']:
                    total += grade
                return round(total / len(student['grades']), 2)
        return None  # Student not found
    
    def getClassAverage(self):
        if len(self.students) == 0:
            return 0
        
        allGrades = []
        for student in self.students:
            allGrades.extend(student['grades'])
        
        if len(allGrades) == 0:
            return 0
        
        total = 0
        for grade in allGrades:
            total += grade
        return round(total / len(allGrades), 2)
    
    def getStudentNames(self):
        return [student['name'] for student in self.students]

# Test
gb = GradeBook()
gb.addStudent("Alice", [90, 85, 88])
gb.addStudent("Bob", [75, 80, 82])
gb.addStudent("Carol", [95, 92, 98])

print(f"Alice's average: {gb.getAverage('Alice')}")
print(f"Bob's average: {gb.getAverage('Bob')}")
print(f"Carol's average: {gb.getAverage('Carol')}")
print(f"Class average: {gb.getClassAverage()}")
print(f"Students: {gb.getStudentNames()}")
```

## Exercise 4: Text Statistics

```python
def textStats(text):
    # Instructor Note: Count chars, words, sentences, avg word length
    charCount = len(text)
    
    # Count words by splitting on whitespace
    words = text.split()
    wordCount = len(words)
    
    # Count sentences by counting . ! ?
    sentenceCount = text.count('.') + text.count('!') + text.count('?')
    
    # Average word length (count letters only, not punctuation)
    totalWordChars = 0
    for word in words:
        # Remove punctuation from word for counting
        cleanWord = ""
        for char in word:
            if char.isalpha():
                cleanWord += char
        totalWordChars += len(cleanWord)
    
    avgWordLength = round(totalWordChars / wordCount, 2) if wordCount > 0 else 0
    
    return {
        'char_count': charCount,
        'word_count': wordCount,
        'sentence_count': sentenceCount,
        'avg_word_length': avgWordLength
    }

# Test
sample = "Hello world! How are you? I am fine."
stats = textStats(sample)
print("Text stats for:", sample)
for key, value in stats.items():
    print(f"  {key}: {value}")

print()
sample2 = "Python is powerful."
stats2 = textStats(sample2)
print("Text stats for:", sample2)
for key, value in stats2.items():
    print(f"  {key}: {value}")
```
