```python
import string

class TextAnalyzer:
    
    def __init__(self, text):
        # Instructor Note: Store original text and cleaned version
        self.originalText = text
        
        # Remove punctuation for word analysis
        self.cleanedText = text.lower()
        for char in string.punctuation:
            self.cleanedText = self.cleanedText.replace(char, "")
    
    def wordCount(self):
        words = self.cleanedText.split()
        return len(words)
    
    def sentenceCount(self):
        count = self.originalText.count('.') + self.originalText.count('!') + self.originalText.count('?')
        return count
    
    def averageWordLength(self):
        words = self.cleanedText.split()
        if len(words) == 0:
            return 0
        
        totalChars = 0
        for word in words:
            totalChars += len(word)
        
        avg = totalChars / len(words)
        return round(avg, 2)
    
    def mostCommonWord(self):
        # Instructor Note: Count word frequencies
        words = self.cleanedText.split()
        
        frequency = {}
        for word in words:
            if word in frequency:
                frequency[word] += 1
            else:
                frequency[word] = 1
        
        if len(frequency) == 0:
            return None
        
        # Find word with max count
        maxWord = None
        maxCount = 0
        for word, count in frequency.items():
            if count > maxCount:
                maxCount = count
                maxWord = word
        
        return maxWord
    
    def longestWord(self):
        # Instructor Note: Find word with most characters (preserve original case)
        # Split by whitespace first
        rawWords = self.originalText.split()
        
        maxWord = ""
        maxLen = 0
        
        for word in rawWords:
            # Remove punctuation from word
            cleanWord = ""
            for char in word:
                if char.isalpha():
                    cleanWord += char
            
            if len(cleanWord) > maxLen:
                maxLen = len(cleanWord)
                maxWord = cleanWord
        
        return maxWord
    
    def analyze(self):
        # Instructor Note: Print formatted report
        print("=" * 30)
        print("TEXT ANALYSIS REPORT")
        print("=" * 30)
        print(f"Word Count: {self.wordCount()}")
        print(f"Sentence Count: {self.sentenceCount()}")
        print(f"Average Word Length: {self.averageWordLength()}")
        
        mostCommon = self.mostCommonWord()
        # Count occurrences
        words = self.cleanedText.split()
        commonCount = words.count(mostCommon) if mostCommon else 0
        print(f"Most Common Word: {mostCommon} (appears {commonCount} times)")
        
        print(f"Longest Word: {self.longestWord()} ({len(self.longestWord())} characters)")
        print("=" * 30)
```

## Test Cases

```python
# Test 1: Simple example
text1 = "Python is great. I love Python!"

analyzer1 = TextAnalyzer(text1)
print(f"Word count: {analyzer1.wordCount()}")
print(f"Sentence count: {analyzer1.sentenceCount()}")
print(f"Avg word length: {analyzer1.averageWordLength()}")
print(f"Most common: {analyzer1.mostCommonWord()}")
print(f"Longest word: {analyzer1.longestWord()}")
print()
analyzer1.analyze()
```

```python
# Test 2: Longer text
text2 = """Python is a powerful programming language. It's easy to learn.
Many programmers love Python. Python is used everywhere!
Machine learning, data science, web development. Python does it all."""

analyzer2 = TextAnalyzer(text2)
analyzer2.analyze()
```

```python
# Test 3: Edge case - single sentence
text3 = "The quick brown fox."

analyzer3 = TextAnalyzer(text3)
analyzer3.analyze()
```
