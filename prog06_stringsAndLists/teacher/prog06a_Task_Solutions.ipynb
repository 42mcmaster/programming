## Exercise 1: Reverse a String

```python
def reverseString(s):
    # Instructor Note: Simplest approach using slicing
    return s[::-1]

# Alternative: manual loop (more like C style)
def reverseString_loop(s):
    result = ""
    for i in range(len(s) - 1, -1, -1):
        result += s[i]
    return result

# Test both
print(reverseString("hello"))
print(reverseString("Python"))
print(reverseString("racecar"))
print(reverseString_loop("hello"))
```

## Exercise 2: Count Vowels

```python
def countVowels(s):
    # Instructor Note: Case-insensitive counting
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

# Alternative using .lower() and .count()
def countVowels_alt(s):
    vowels = "aeiou"
    s_lower = s.lower()
    count = 0
    for v in vowels:
        count += s_lower.count(v)
    return count

# Test
print(countVowels("hello"))  # 2: e, o
print(countVowels("beautiful"))  # 5: e, a, u, i, u
print(countVowels("xyz"))  # 0
print(countVowels("AEIOUaeiou"))  # 10
print(countVowels_alt("HELLO"))  # 2
```

## Exercise 3: Check Palindrome

```python
def isPalindrome(s):
    # Instructor Note: Remove spaces and convert to lowercase
    cleaned = s.replace(" ", "").lower()
    return cleaned == cleaned[::-1]

# Alternative: explicit comparison
def isPalindrome_alt(s):
    cleaned = s.replace(" ", "").lower()
    for i in range(len(cleaned) // 2):
        if cleaned[i] != cleaned[-(i + 1)]:
            return False
    return True

# Test
print(isPalindrome("racecar"))  # True
print(isPalindrome("hello"))  # False
print(isPalindrome("A man a plan a canal Panama"))  # True
print(isPalindrome("Madam"))  # True
print(isPalindrome("Python"))  # False
print(isPalindrome_alt("racecar"))  # True
```

## Exercise 4: Extract Initials

```python
def getInitials(name):
    # Instructor Note: Split words, take first char of each, uppercase
    words = name.split()
    initials = ""
    for word in words:
        if len(word) > 0:
            initials += word[0].upper()
    return initials

# Alternative using list comprehension
def getInitials_comp(name):
    return "".join([word[0].upper() for word in name.split() if word])

# Test
print(getInitials("John Quincy Adams"))  # JQA
print(getInitials("ryan mcmaster"))  # RM
print(getInitials("Mary Jane Watson"))  # MJW
print(getInitials_comp("John Quincy Adams"))  # JQA
```

## Exercise 5: Caesar Shift

```python
def caesarShift(s, shift):
    # Instructor Note: Only shift lowercase a-z, wrap with modulo
    result = ""
    
    for char in s:
        if 'a' <= char <= 'z':
            # Convert char to 0-25, apply shift, wrap, convert back
            charIndex = ord(char) - ord('a')
            newIndex = (charIndex + shift) % 26
            newChar = chr(ord('a') + newIndex)
            result += newChar
        else:
            # Leave non-lowercase chars unchanged
            result += char
    
    return result

# Test
print(caesarShift("hello", 1))  # ifmmp
print(caesarShift("xyz", 3))  # abc
print(caesarShift("hello 123", 5))  # mjqqt 123
print(caesarShift("abc", 26))  # abc (full rotation)
print(caesarShift("hello", 0))  # hello
```

## Bonus Challenge: Remove Vowels

```python
def removeVowels(s):
    # Instructor Note: Multiple approaches
    vowels = "aeiouAEIOU"
    result = ""
    for char in s:
        if char not in vowels:
            result += char
    return result

# Alternative using list comprehension
def removeVowels_comp(s):
    vowels = "aeiouAEIOU"
    return "".join([char for char in s if char not in vowels])

# Alternative using replace
def removeVowels_replace(s):
    for vowel in "aeiouAEIOU":
        s = s.replace(vowel, "")
    return s

# Test all three
print(removeVowels("Hello World"))  # Hll Wrld
print(removeVowels("Programming"))  # Prgrmmng
print(removeVowels("aeiou"))  # (empty)
print(removeVowels_comp("Hello World"))  # Hll Wrld
print(removeVowels_replace("Hello World"))  # Hll Wrld
```
